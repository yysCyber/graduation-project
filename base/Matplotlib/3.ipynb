# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

### 1

```python
s1 = Series(np.random.randn(1000)).cumsum()
s1
```

```python
plt.show(s1.plot())
```

### 2

```python
s2 = Series(np.random.randn(10)).cumsum()
s2
```

```python
plt.show(s2.plot(kind='bar'))
```

### 3

```python
s3 = Series(np.random.randn(100)).cumsum()
s3
```

```python
plt.show(s3.plot(kind='line', grid=True))
```

### 4

```python
s4 = Series(np.random.randn(100)).cumsum()
s4
```

```python
img_1 = s4.plot(kind='line', grid='True', label='s4')
plt.legend()
plt.show(img_1)
```

### 5

```python
s5 = Series(np.random.randn(50)).cumsum()
s5
```

```python
img_2 = s5.plot(kind='line', grid='True', label='s5', style='r--', title='This is series 5')
plt.legend()
plt.show(img_2)
```

### 6

```python
s6 = Series(np.random.randn(30)).cumsum()
s6
```

```python
s7 = Series(np.random.randn(30)).cumsum()
s7
```

```python
img_2 = s6.plot(kind='line', label='s6', style='r-', grid=True, title='s6 & s7')
img_3 = s7.plot(kind='line', label='s7', style='b--', grid=True)
plt.legend()
plt.show(img_2, img_3)
```

### 7

```python
s8 = Series(np.random.randn(100).cumsum())
s9 = Series(np.random.randn(100).cumsum())
```

```python
figure, ax = plt.subplots(2, 1)
ax[0].plot(s8, 'r-')
ax[1].plot(s9, 'b--')
plt.show()
```

### 8

```python
s10 = Series(np.random.randn(100).cumsum())
s11 = Series(np.random.randn(100).cumsum())
```

```python
figure, ax = plt.subplots(2, 1)
s10.plot(ax=ax[0], style='r-', label='s10', legend=True, grid=True)
s11.plot(ax=ax[1], style='b:', label='s11', legend=True, grid=True)
plt.show()
```
