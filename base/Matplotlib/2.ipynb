# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1

```python
x1 = np.linspace(0.0, 5.0)
x1
```

```python
y1 = np.sin(x1 * np.pi)
y1
```

```python
y2 = np.sin(x1 * np.pi * 2)
y2
```

```python
plt.plot(x1, y1, 'r-', label='y=sin(πx)')
plt.plot(x1, y2, 'b--', label='y=sin(2πx)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### 2

```python
plt.xlabel('x')

plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'r-')
plt.ylabel('y1.1')

plt.subplot(2, 2, 2)
plt.plot(x1, y1, 'b-')
plt.ylabel('y1.2')

plt.subplot(2, 2, 3)
plt.plot(x1, y2, 'b--')
plt.ylabel('y2.1')


plt.subplot(2, 2, 4)
plt.plot(x1, y2, 'r--')
plt.ylabel('y2.2')

plt.show()
```

```python
plt.xlabel('x')

plt.subplot(221)
plt.plot(x1, y1, 'r-')
plt.ylabel('y1.1')

plt.subplot(222)
plt.plot(x1, y1, 'b-')
plt.ylabel('y1.2')

plt.subplot(223)
plt.plot(x1, y2, 'b:')
plt.ylabel('y2.1')


plt.subplot(224)
plt.plot(x1, y2, 'r-.')
plt.ylabel('y2.2')

plt.show()
```
