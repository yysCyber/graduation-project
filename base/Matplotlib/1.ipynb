### 4

```python
# y = 2 * x
l4 = [0, 1, 2, 3, 4, 5]
l5 = [0, 2, 4, 6, 8, 10]
```

```python
# y = 3 * x + 2
l6 = [0, 1, 2, 3, 4, 5]
l7 = [2, 5, 8, 11, 14, 17]
```

```python
plt.plot(l4, l5, 'r-', l6, l7, 'b--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('demo1')
```

### 5

```python
s1 = np.arange(0.0, 2.1, 0.1)
s1
```

```python
s2 = np.cos(s1 * np.pi)
s2
```

```python
s3 = np.sin(s1 * np.pi)
s3
```

```python
plt.show(plt.plot(s1, s2, 'r-', s1, s3, 'b--'))
```

### 6

```python
s4 = np.arange(0.0, 2*np.pi+1, np.pi/2)
s4
```

```python
s5 = np.sin(s4)
s5
```

```python
plt.show(plt.plot(s4, s5, 'r-'))
```

### 7

```python
plt.plot(l4, l5, 'r-', label='y=2*x')
plt.plot(l6, l7, 'b--', label='y=3*x+2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('demo2')
plt.legend()
```
