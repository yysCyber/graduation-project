```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
```

### 1

```python
s1 = Series(np.random.randn(1000))
s1
```

```python
plt.hist(s1)
plt.show()
```

### 3

```python
s3 = Series(np.random.randn(1000))
s3
```
