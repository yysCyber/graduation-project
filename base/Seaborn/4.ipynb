# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
```

### 1

```python
x1 = np.linspace(0, 14, 100)
```

```python
y1_1 = np.sin(x1)
y1_2 = np.sin(x1 + 2) * 1.25
```

```python
def sin_plot():
    plt.plot(x1, y1_1)
    plt.plot(x1, y1_2)
```

```python
sin_plot()
```

### 2

```python
sns.set()
```

```python
sin_plot()
```
