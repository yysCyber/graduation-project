# Pandas

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

### 1

```python
s1 = Series(np.random.randn(10))
s1
```

```python
s2 = s1.sort_values()
s2
```

```python
s3 = s1.sort_values(ascending=False)
s3
```

### 2

```python
s4 = s1.sort_index()
s4
```

```python
s5 = s1.sort_index(ascending=False)
s5
```

### 3

```python
df1 = DataFrame(np.random.randn(40).reshape(8, 5), columns=['A', 'B', 'C', 'D', 'E'])
df1
```

```python
df2 = df1.sort_values('A')
df2
```

```python
df3 = df1.sort_values('A', ascending=False)
df3
```
