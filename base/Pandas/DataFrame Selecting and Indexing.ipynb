```python
sub_data_source = data_source[['movie_title', 'director_name', 'imdb_score']]
```

```python
sub_data_source.head()
```

```python
# 6、基于索引选取行/列
# iloc 的“[]”中的第一个参数是用来过滤行的，第二个参数是用来过滤列的，这两者可以使用“切片”的方法
# 注意，这个参数是索引，索引所对于的是此时此刻当前的 DataFrame 而言的

# 选取 10-19 行
sub_data_source.iloc[10:20, :]
```

```python
sub_data_source.iloc[[10, 11, 13, 15], :]
```

```python
sub_data_source.iloc[10:20, 0:3]
```

```python
sub_data_source.iloc[10:20, [1, 2]]
```

<hr>

```python
# 7、基于标签选取行/列
# 对于行而言，label 是指最左侧的所添加出来的列的值（在之前的文档中，将这个称为“行索引”，实际上准确来讲应该是“行标签”）；
# 对于列而言，label 就是列名

sub_data_source.loc[0:4, :]
```

```python
sub_data_source.loc[0:4, 'movie_title']
```

```python
sub_data_source.loc[0:4, ['movie_title', 'imdb_score']]
```
