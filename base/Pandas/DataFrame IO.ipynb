# Pandas DataFrame I/O

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

### 剪切板

```python
# 从“剪切板”中读取数据创建 DataFrame
df1 = pd.read_clipboard()

# dataFrame.to_clipboard()：将 dataFrame 写入“剪切板”
```

```python
df1
```

### CSV 文件

```python
# 将 DataFrame 中的内容写入 .csv 文件
```

```python
df2 = df1
df2.to_csv('test/df2_1.csv')
```

```python
# 因为 DataFrame 默认会自动在最前面追加一列自0开始的索引，但是生成 .csv 文件的时候不希望要这一列，可以使用下面的
df2.to_csv('test/df2_2.csv', index=False)
```
