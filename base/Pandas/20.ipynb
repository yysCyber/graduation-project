# Pandas

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

### 1

```python
arr1 = np.random.randint(low=40, high=101, size=20)
arr1
```

```python
# (0, 59] 60分以下 不及格
# (59, 69] 60~69 及格
# (69, 79] 70~79 一般
# (79, 89] 80~89 良好
# (89, 100] 90~100 优秀
bins_1 = [0, 59, 69, 79, 89, 100]
```

```python
cut_result_1 = pd.cut(x=arr1, bins=bins_1)
cut_result_1
```

```python
pd.value_counts(cut_result_1)
```

### 2

```python
df1 = pd.read_csv('datasource\\my_grade.csv')
df1
```

```python
df1 = df1[['课程编号', '课程名称', '成绩']]
df1 = df1.rename(index=Series(np.arange(1, 64, 1)),columns={'课程编号': 'No', '课程名称': 'Name', '成绩': 'Score'})
df1
```

```python
bins_2 = [0, 59, 69, 79, 89, 100]
```

```python
df1['Category'] = pd.cut(x=df1['Score'], bins=bins_2)
```

```python
df1
```

```python
df1.to_csv('datasource\\my_grade_1.csv', index=False)
```
