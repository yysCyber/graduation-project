### 3

```python
df3 = df1.copy()
df3
```

```python
def check_is_delayed(arr_delay):
    if arr_delay > 0:
        return 'Yes'
    else:
        return 'No'
```

```python
df3['is_delayed'] = df3['arr_delay'].apply(check_is_delayed)
```

```python
df3
```

### 4

```python
df4 = df1.copy()
df4
```

```python
df4['is_delayed'] = df4['arr_delay'].apply(lambda x: x>0)
df4
```

```python
df4.head()
```

```python
s1 = df4['is_delayed'].value_counts()
s1
```

```python
(s1[True] / (s1[True] + s1[False])) * 100
```
