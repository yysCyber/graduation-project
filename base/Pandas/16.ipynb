### 3

```python
df2 = df.copy()
```

```python
df2.head()
```

```python
df2['data'][0].split()
```

```python
def split_data(data):
    items = data.split()
    return Series([items[1], items[3], items[5]])
```

```python
df2_temp = df2['data'].apply(split_data)
```

```python
df2_temp
```

```python
df2_temp = df2_temp.rename(columns={0: 'symbol', 1: 'seqno', 2: 'price'})
```

```python
df2_temp
```

```python
df3 = df2.combine_first(df2_temp).drop(columns='data')
```

```python
df3 = df3.reindex(columns=['time', 'symbol', 'seqno', 'price'])
```

```python
df3
```

```python
df3.to_csv('datasource\\duplicate_demo.csv', index=False)
```
